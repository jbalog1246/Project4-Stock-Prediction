**Based on the growth of the dividends, TGT, ABBV, JPM, JNJ, and PEP have the best trend. Now I will compare the 5 stocks on their average price over 20 days and as well as 200 days, and then choose the top 3 to invest in.**

```python
start_date = '2020-01-01'
end_date = '2021-12-01'

Target = yf.download('TGT', start_date, end_date)
AbbVie = yf.download('ABBV', start_date, end_date)
JPMorgan = yf.download('JPM', start_date, end_date)
Johnson = yf.download('JNJ', start_date, end_date)
Pepsi = yf.download('PEP', start_date, end_date)
```

```python
Target['MA50'] = Target['Open'].rolling(20).mean()
Target['MA200'] = Target['Open'].rolling(200).mean()
Target['Open'].plot(figsize = (14,7))
Target['MA50'].plot()
Target['MA200'].plot()
```

```python
AbbVie['MA50'] = AbbVie['Open'].rolling(20).mean()
AbbVie['MA200'] = AbbVie['Open'].rolling(200).mean()
AbbVie['Open'].plot(figsize = (14,7))
AbbVie['MA50'].plot()
AbbVie['MA200'].plot()
```

```python
JPMorgan['MA50'] = JPMorgan['Open'].rolling(20).mean()
JPMorgan['MA200'] = JPMorgan['Open'].rolling(200).mean()
JPMorgan['Open'].plot(figsize = (14,7))
JPMorgan['MA50'].plot()
JPMorgan['MA200'].plot()
```

```python
Johnson['MA50'] = Johnson['Open'].rolling(20).mean()
Johnson['MA200'] = Johnson['Open'].rolling(200).mean()
Johnson['Open'].plot(figsize = (14,7))
Johnson['MA50'].plot()
Johnson['MA200'].plot()
```

```python
Pepsi['MA50'] = Pepsi['Open'].rolling(20).mean()
Pepsi['MA200'] = Pepsi['Open'].rolling(200).mean()
Pepsi['Open'].plot(figsize = (14,7))
Pepsi['MA50'].plot()
Pepsi['MA200'].plot()
```

**Looking at the 5 stocks, the 3 that are performing the best are Target, JPMorgan and Pepsi. I will by 10 shares of each of the stocks for this week.**
