```python
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import yfinance as yf
```

```python
start_date = '2020-01-01'
end_date = '2021-11-22'

Penn = yf.download('PENN', start_date, end_date)
Microsoft = yf.download('MSFT', start_date, end_date)
Ford = yf.download('F', start_date, end_date)
NVIDIA = yf.download('NVDA', start_date, end_date)
Coinbase = yf.download('COIN', start_date, end_date)
SoFi = yf.download('SOFI', start_date, end_date)
```

Now I will collect the average price over 20 days and as well as 200 days.

```python
Penn['MA50'] = Penn['Open'].rolling(20).mean()
Penn['MA200'] = Penn['Open'].rolling(200).mean()
Penn['Open'].plot(figsize = (14,7))
Penn['MA50'].plot()
Penn['MA200'].plot()
```

```python
Microsoft['MA50'] = Microsoft['Open'].rolling(20).mean()
Microsoft['MA200'] = Microsoft['Open'].rolling(200).mean()
Microsoft['Open'].plot(figsize = (14,7))
Microsoft['MA50'].plot()
Microsoft['MA200'].plot()
```

```python
Ford['MA50'] = Ford['Open'].rolling(20).mean()
Ford['MA200'] = Ford['Open'].rolling(200).mean()
Ford['Open'].plot(figsize = (14,7))
Ford['MA50'].plot()
Ford['MA200'].plot()
```

```python
NVIDIA['MA50'] = NVIDIA['Open'].rolling(20).mean()
NVIDIA['MA200'] = NVIDIA['Open'].rolling(200).mean()
NVIDIA['Open'].plot(figsize = (14,7))
NVIDIA['MA50'].plot()
NVIDIA['MA200'].plot()
```

```python
Coinbase['MA50'] = Coinbase['Open'].rolling(20).mean()
Coinbase['MA200'] = Coinbase['Open'].rolling(200).mean()
Coinbase['Open'].plot(figsize = (14,7))
Coinbase['MA50'].plot()
Coinbase['MA200'].plot()
```

```python
SoFi['MA50'] = SoFi['Open'].rolling(20).mean()
SoFi['MA200'] = SoFi['Open'].rolling(200).mean()
SoFi['Open'].plot(figsize = (14,7))
SoFi['MA50'].plot()
SoFi['MA200'].plot()
```
