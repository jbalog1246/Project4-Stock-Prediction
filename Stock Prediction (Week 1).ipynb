```python
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import yfinance as yf
```

```python
start_date = '2020-01-01'
end_date = '2021-11-18'

Apple = yf.download('AAPL', start_date, end_date)
Microsoft = yf.download('MSFT', start_date, end_date)
Google = yf.download('GOOG', start_date, end_date)
IBM = yf.download('IBM', start_date, end_date)
Amazon = yf.download('AMZN', start_date, end_date)
Facebook = yf.download('FB', start_date, end_date)
```

```python
Apple['Open'].plot(label = 'AAPL', figsize = (15,7))
Microsoft['Open'].plot(label = "MSFT")
Google['Open'].plot(label = 'GOOGL')
IBM['Open'].plot(label = "IBM")
Amazon['Open'].plot(label = 'AMZN', figsize = (15,7))
Facebook['Open'].plot(label = "FB")

plt.title('Stock Prices')
plt.legend()
```

The graph shown above are all of the stocks prices. Google and Amazon currently have the highest prices for their stock.

```python
Apple['Volume'].plot(label = 'AAPL', figsize = (15,7))
Microsoft['Volume'].plot(label = "MSFT")
Google['Volume'].plot(label = 'GOOGL')
IBM['Volume'].plot(label = "IBM")
Amazon['Volume'].plot(label = "AMZN")
Facebook['Volume'].plot(label = "FB")

plt.title('How many times the stock was traded')
plt.legend()
```

This graph shows how many times each stock was traded. As seen, Apple is highly traded followed by Microsoft/Facebook. Comparing the graphs, Amazon and Google seem to be doing good since they are steadily increasing their value. As for Apple, it has a steady price but is traded more highly than other stocks
