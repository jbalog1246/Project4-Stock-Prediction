```python
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import yfinance as yf
```

```python
start_date = '2020-01-01'
end_date = '2021-12-03'

Target = yf.download('TGT', start_date, end_date)
Greif = yf.download('GEF', start_date, end_date)
AbbVie = yf.download('ABBV', start_date, end_date)
JPMorgan = yf.download('JPM', start_date, end_date)
Johnson = yf.download('JNJ', start_date, end_date)
Iron = yf.download('IRM', start_date, end_date)
Pepsi = yf.download('PEP', start_date, end_date)
Discover = yf.download('DFS', start_date, end_date)
Cisco = yf.download('CSCO', start_date, end_date)
Att = yf.download('T', start_date, end_date)
```

I will first find the opening price for 5 of the stocks and then do the same for the other 5 stocks in a different graph.

```python
Target['Open'].plot(label = 'TGT', figsize = (15,7))
Greif['Open'].plot(label = "GEF")
AbbVie['Open'].plot(label = 'ABBV')
JPMorgan['Open'].plot(label = "JPM")
Johnson['Open'].plot(label = 'JNJ', figsize = (15,7))

plt.title('Stock Prices')
plt.legend()
```

```python
Iron['Open'].plot(label = 'IRM', figsize = (15,7))
Pepsi['Open'].plot(label = "PEP")
Discover['Open'].plot(label = 'DFS')
Cisco['Open'].plot(label = "CSCO")
Att['Open'].plot(label = 'T', figsize = (15,7))

plt.title('Stock Prices')
plt.legend()
```

Now I will look at the volume the stocks were traded at.

```python
Target['Volume'].plot(label = 'TGT', figsize = (15,7))
Greif['Volume'].plot(label = "GEF")
AbbVie['Volume'].plot(label = 'ABBV')
JPMorgan['Volume'].plot(label = "JPM")
Johnson['Volume'].plot(label = 'JNJ', figsize = (15,7))

plt.title('How many times the stocks were traded')
plt.legend()
```

```python
Iron['Volume'].plot(label = 'IRM', figsize = (15,7))
Pepsi['Volume'].plot(label = "PEP")
Discover['Volume'].plot(label = 'DFS')
Cisco['Volume'].plot(label = "CSCO")
Att['Volume'].plot(label = 'T', figsize = (15,7))

plt.title('How many times the stocks were traded')
plt.legend()
```

Now I will multiply the opening price * volume.

```python
Target['MarktCap'] = Target['Open'] * Target['Volume']
Greif['MarktCap'] = Greif['Open'] * Greif['Volume']
AbbVie['MarktCap'] = AbbVie['Open'] * AbbVie['Volume']
JPMorgan['MarktCap'] = JPMorgan['Open'] * JPMorgan['Volume']
Johnson['MarktCap'] = Johnson['Open'] * Johnson['Volume']

Target['MarktCap'].plot(label = 'TGT', figsize = (15,7))
Greif['MarktCap'].plot(label = "GEF")
AbbVie['MarktCap'].plot(label = 'ABBV')
JPMorgan['MarktCap'].plot(label = 'JPM')
Johnson['MarktCap'].plot(label = "JNJ")

plt.title('Opening Price * Volume')
plt.legend()
```

```python
Iron['MarktCap'] = Iron['Open'] * Iron['Volume']
Pepsi['MarktCap'] = Pepsi['Open'] * Pepsi['Volume']
Discover['MarktCap'] = Discover['Open'] * Discover['Volume']
Cisco['MarktCap'] = Cisco['Open'] * Cisco['Volume']
Att['MarktCap'] = Att['Open'] * Att['Volume']

Iron['MarktCap'].plot(label = 'IRM', figsize = (15,7))
Pepsi['MarktCap'].plot(label = "PEP")
Discover['MarktCap'].plot(label = 'DFS')
Cisco['MarktCap'].plot(label = "CSCO")
Att['MarktCap'].plot(label = 'T')

plt.title('Opening Price * Volume')
plt.legend()
```

Now I will take the top 5 companies (TGT, JPM, PEP, CSCO, and T) and research their average price over 20 days and 200 days.

```python
Target['MA50'] = Target['Open'].rolling(20).mean()
Target['MA200'] = Target['Open'].rolling(200).mean()
Target['Open'].plot(figsize = (14,7))
Target['MA50'].plot()
Target['MA200'].plot()
```

```python
JPMorgan['MA50'] = JPMorgan['Open'].rolling(20).mean()
JPMorgan['MA200'] = JPMorgan['Open'].rolling(200).mean()
JPMorgan['Open'].plot(figsize = (14,7))
JPMorgan['MA50'].plot()
JPMorgan['MA200'].plot()
```

```python
Pepsi['MA50'] = Pepsi['Open'].rolling(20).mean()
Pepsi['MA200'] = Pepsi['Open'].rolling(200).mean()
Pepsi['Open'].plot(figsize = (14,7))
Pepsi['MA50'].plot()
Pepsi['MA200'].plot()
```

```python
Cisco['MA50'] = Cisco['Open'].rolling(20).mean()
Cisco['MA200'] = Cisco['Open'].rolling(200).mean()
Cisco['Open'].plot(figsize = (14,7))
Cisco['MA50'].plot()
Cisco['MA200'].plot()
```

```python
Att['MA50'] = Att['Open'].rolling(20).mean()
Att['MA200'] = Att['Open'].rolling(200).mean()
Att['Open'].plot(figsize = (14,7))
Att['MA50'].plot()
Att['MA200'].plot()
```
